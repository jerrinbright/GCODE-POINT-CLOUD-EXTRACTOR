```
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as pp
from matplotlib.animation import FuncAnimation
pp.style.use('seaborn-pastel')
import matplotlib.pyplot as plt
import matplotlib.animation as animation

Xdot=[]
Ydot=[]
Zdot=[]
file=open('a.txt','r')
for line in file:
    if line[0]=='G':
        words=line.split()
        for word in words:
            if word[0]=='X':
                w=word[1:len(word)]
                Xdot.append(float(w))
            elif word[0]=='Y':
                w=word[1:len(word)]
                Ydot.append(float(w))
            elif word[0]=='E':
                w=word[1:len(word)]
                Zdot.append(w)
for i in range(len(Zdot)):
    if Zdot[i]=='':
        Zdot[i]='0'
for i in range(len(Zdot)):
    Zdot[i]=float(Zdot[i])


d = np.array(Xdot)
d1 = np.array(Zdot)
d2 = np.array(Ydot)

print(Xdot)
print(Ydot)
print(Zdot)
```
